```python
import warnings

import pandas as pd
from pandas.core.common import SettingWithCopyWarning
warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)

import numpy as np
import datetime

pd.set_option("display.max_columns", None)
#pd.set_option("display.max_rows", None)
#pd.options.display.max_rows = 50
```

### Combined df having 3 subwindows

```python
path1 = "/Users/niharawarawita/Desktop/MSc Project/Data and code/Lili_Experiment/combined_stats_lili_3subwindows.csv"
df1 = pd.read_csv(path1)
df1.head()
```

```python
path2 = "/Users/niharawarawita/Desktop/MSc Project/Data and code/2022_Experiment/combined_stats_2022_3subwindows.csv"
df2 = pd.read_csv(path2)
df2.head()
```

```python
# Look at the participant ids for the 2 dfs
print(f'Participant ids for df1 = {sorted(df1.participant_id.unique())}')
print(f'Participant ids for df2 = {sorted(df2.participant_id.unique())}')
```

```python
# Look at the clothes ids for the 2 dfs
print(f'Clothes ids for df1 = {sorted(df1.clothes_id.unique())}')
print(f'Clothes ids for df2 = {sorted(df2.clothes_id.unique())}')

# Change the clothes id for df2
df2['clothes_id'] = df2['clothes_id'] + 30

print(f'New clothes ids for df2 = {sorted(df2.clothes_id.unique())}')
```

```python
# Look at the property ids for the 2 dfs
print(f'Property ids for df1 = {sorted(df1.property_id.unique())}')
print(f'Property ids for df2 = {sorted(df2.property_id.unique())}')

# Change the property ids for df2
df2['property_id'] = df2['property_id'] - 9

print(f'New property ids for df2 = {sorted(df2.property_id.unique())}')
```

```python
print(f'df1 should have {8*6*6} unique interaction ids but only have {df1.interaction_id.nunique()} unique ones')
print(f'df2 should have {6*6*6} unique interaction ids and have {df2.interaction_id.nunique()} unique ones')
```

```python
# Add a new column for the interaction id for df1
df1.insert(5, "new_interaction_id", None)
df1['new_interaction_id'] = df1.groupby(['participant_id', 'clothes_id', 'property_id'], sort=False).ngroup() + 1
df1.drop('interaction_id', axis=1, inplace=True)
df1.rename(columns={'new_interaction_id': 'interaction_id'}, inplace=True)

# Change the property ids for df2
df2['interaction_id'] = df2['interaction_id'] + 181
```

```python
print(f'df1 should have {8*6*6} unique interaction ids and have {df1.interaction_id.nunique()} unique ones')
print(f'df2 should have {6*6*6} unique interaction ids and have {df2.interaction_id.nunique()} unique ones')
```

```python
# Combine the 2 dfs
combined_df = pd.concat([df1, df2], axis=0)
combined_df.head()
```

```python
print(f'Shape of df1 : {df1.shape}')
print(f'Shape of df2 : {df2.shape}')
print(f'Shape of combined df : {combined_df.shape}')
```

```python
print(f'Participant ids for combined df = {sorted(combined_df.participant_id.unique())}')
print(f'Clothes ids for combined df = {sorted(combined_df.clothes_id.unique())}')
print(f'Property ids for combined df = {sorted(combined_df.property_id.unique())}')
```

```python
# Save the created dataset
write_path = "/Users/niharawarawita/Desktop/MSc Project/Data and code/complete_dataset_3subwindows.csv"
combined_df.to_csv(write_path, index = None)
```
